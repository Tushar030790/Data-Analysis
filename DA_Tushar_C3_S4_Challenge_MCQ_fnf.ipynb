```python
Q={"Number of Continents on Earth":"1:Seven 2:Four 3:Five 4:Six Enter S to skip to the next the question", 
   "Which country successfully landed a man on moon":"1:USA 2:Germany 3:France 4:England Enter S to skip to the next the question",
    "Which country is not present in the continent of North America":"1:Mexico 2:USA 3:Cannada 4:Brazil Enter S to skip to the next the question",
    "Which is the American National Bird":"1:Parrot 2:Crane 3:Eagle 4:Swan Enter S to skip to the next the question",
   "Which country shares the longest border with USA":"1:Mexico 2:Brazil 3:Cannada 4:Argentina Enter S to skip to the next the question",
   "Which is the financial capital of USA":"1:Sanfrancisco 2:Boston 3:Los Angeles 4:NYC Enter S to skip to the next the question",
   "Which of the following space agencies belongs to the USA":"1:BNSC 2:ISRO 3:DLR 4:NASA Enter S to skip to the next the question",
   "Which is the second most spoken language in USA":"1:English 2:Spanish 3:Mandarin 4:Italian Enter S to skip to the next the question",
   "Which is the capital city of USA":"1:NYC 2:Washington DC 3:Austin 4:San Jose Enter S to skip to the next the question" ,
   "Mt Rushmore has the sclupture of":"1:Former US Presidents 2:Hollywood stars 3:Military Personel 4:Scientists Enter S to skip to the next the question" 
  } # Declaring a Dictionary with questions as keys and options as values 
# The above dictionary Q contains the questions and their corresponding answer options.

A= list(Q.values()) # creating a list with all the questions as elements 
CA=[1,1,4,3,3,4,4,2,2,1] # List of correct options 
opt=["1","2","3","4","s","S"] # List of all possible options 
UA=[] # Empty list to store the answers entered by user 
myit = iter(A) # an iterator to iterate the statement one by one 
for i in Q: # for loop to iterate the statement
    print(i) #  Print Questions
    a=next(myit)  # print options 
    print(a) # Print 
    #print("\n")
    ans = input("Enter the one option between 1 to 4 and s to Skip:") # taking input from the user 
    # if user entered wrong input other than 1,2,3,4 or s it will ask to re-enter the the option 1-4 or s
    # Until user not entered option 1-4 or s it will not proceed further 
    while ans not in opt:  # While loop to iterate the statements
        print("invalid Input")  # if entered input other than 1-4 or s it is invalid input
        ans=input("Please Enter the one option between 1 to 4 and s to Skip:") # ask user to reenter the option
         # IF user again enter wrong input than we continue the while loop and print invalid input and ask to
            # again enter the options 
        if ans not in opt: 
            continue
        else: # if user entered the correct input then we will break the loop 
            break
    if ans=="s" or ans=="S": # Conditional Statement to check if the user enter "S" or "s"
        UA.append("s") #  if true than add 0 to User answer list
    else:
        UA.append(ans) # if false then  add answer to the to the UA List
    print("\n")

crt=0 # variable to count correct questions
incrt=0 # variable to count incorrect question
No_atmp=0
for i in range (0, len(CA)): # for loop iterate the statemen
    if UA[i]=="s":
        No_atmp=No_atmp+1
    elif UA[i]==str(CA[i]): # Conditional Statement to check if the answer entered by user is matching with correct answer
        crt=crt+1 # if matches then increament the correct count 
    else:
        incrt=incrt+1 # if not then increamne the incorrect count 
print("Correct Answer",crt) # Printing number of correct answer 
print("InCorrect Answer",incrt) # Printing incorrect answer
print("No attempt", No_atmp)
Neg_Sc=incrt*0.25
Total_Score=crt*2-Neg_Sc
print("Total Negetive Marks: ", Neg_Sc)
Per=Total_Score*100/20 # Calculating the percentage 

print("You have Scored {}/20 and {}%".format(Total_Score,Per)) # Print the total score and percentage. 
```
