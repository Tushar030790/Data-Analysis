```python
class color: # A user defined class to print in Bold and underlines 
   BOLD = '\033[1m'
   UNDERLINE = '\033[4m'
   END = '\033[0m'
```

```python
# A user defined function to print Bill and calculate the items cost and total amount with delivery charge
# To calculate the Total amount we need the delivery charge and to print the bill we required a nested list
# So the function take two argument
def bill (l,m):    # Defination of function 
    Total_item_cost=0 # intialise a variable to calculate the total Bill Amount with delivery charge 
    c=1 # Counter variable 
    print("%45s" %(color.BOLD + "BILL/INVOICE" + color.END)) #  print in bold letter
    print("%1s %15s %25s %10s" %("S.no", "items", "Quantity", "Cost/item")) # Statment to print the header 
    for i in range (0, len(l)): # for loop for printing nested list
        print("%2s %20s %20s %10s" %(c,l[i][0],l[i][1],l[i][2]))
        Total_item_cost= Total_item_cost+l[i][2]
        c=c+1 # increament in counter variable 
    Total_amount=Total_item_cost+m # calculate the total amount that is amount of all items + delivery charge 
    print("Total Item cost is: ${}" .format(Total_item_cost)) # printing total item cost 
    print("\n") 
    print("Total Bill Amount = Total item cost + Delivery charge = ${}" .format(Total_amount)) # Printing total amount
```

```python
def delivery_charges(dis): # function to return the delivery charge 
    if dis<=5: # if distance is less than 5 than delivery charge is 10$
        return 10 # return 10
    elif dis>5 and dis<=10: # if distance is less than 10 tham delivery charge is $12
        return 12
    elif dis>10 and dis<=15:# if distance is less than 15 tham delivery charge is $14
        return 14
    else:
        return 20 # if distance is above  15 tham delivery charge is $14
```

```python
# function to update the quantity in the store
# it take two arguments first is nested dictionary and list if quantities 
def update_items(u, l): 
    k=0 # counter variable 
    for i,j in u.items(): # for loop to iterate over the nested dictionary
        u[i]["quantity"]=u[i]["quantity"]-l[k] # update the quantity with the current stoke
        k=k+1 # increament the counter variable 
```

```python
# declaring the nested dictionary to store the item in the store 
d ={"Biscuit:a pack of 20":{"quantity": 5, "cost":2.5},
    "Cereals:2 pound/pack":{"quantity": 10, "cost":3.5},
    "Chicken:1 pound/pack":{"quantity": 20, "cost":13.5},
    "Oats:4 pound/pack   ":{"quantity": 20, "cost":3.9},
    "Rice:2 pound/pack   ":{"quantity": 30, "cost":4.5}}

def dis_items(d): # finction to display the store items 
    b=1   # counter variable to print serial number 
    print("%45s" %(color.BOLD + "ESSENTIAL ITEMS AVAILABLE" + color.END)) #  print in bold letter
    print("%1s %15s %25s %10s" %("S.no", "items", "Quantity", "Cost/item in $")) # print the header
    for i,j in d.items(): # for loop to iterate over the nested dictionary
        print("%5s %25s" %(b, i), end=" ") # print serial number and keys 
        for k,l in j.items(): # for the to iterate over the inner dictionary
            print("%10s" %(l),end="    ") # print values of inner dictionary
        b=b+1 # increament the counter
        print("\n")
    print("\n")      
while True: # infinite loop 
    dis_items(d) # calling the display items function to print the store items, quantity and cost/item
    quantity=[] # Empty list to store number of quantities
    bill_item=[] # Empty list to store the ordered items and quantity in nested list
    cost_item=[] # Empty list to store the cost of items 
    for i,j in d.items(): # for loop to iterate over the loop
        q=int(input("How much packet of {} are required?" .format(i))) # input from user to get the quantity
        quantity.append(q) # append it to the quantity list
        if q>d[i]["quantity"]: # quantity is more than avaible quantity than we need to enter the quantity again 
            print("Out of Stock") # Print the item is out of stock please enter less quantity
            continue
        else:
            cost = q*d[i]["cost"] # if the less then aviable quantity than we calculate the cost of item 
            cost_item.append(cost) # append the cost to cost_item list
            bill_item.append([i,q,cost]) # add the keys, quantity and cost to cart_items
            
    print("\n")
    update_items(d,quantity) # calling update function to update the quantity 
    print(color.BOLD + "Fill in your Delivery details" + color.END) # print in Bold
    print("\n")

    name=input("Enter your name: ")  # input name 
    add=input("Enter your Address: ") # inout delivery address
    dis=int(input("Distance from store in miles 0-5/5-10/10-15/Above 15: ")) # input distance from the store of delivery address
    

    del_charges= delivery_charges(dis) # delivery charge is return by the delivery function and store to del_charge
    bill(bill_item,del_charges) # calling bill function with bill_item lested dictionary and del_charges 
        
    dscn=input("Do you want to continue Shopping? y/n: ") # decision of user if they want to continue shopping 
    if dscn.lower()=="y": # if true then continue the infinite loop  
        continue # continue statment 
    else: # else 
        break # break the loop 
 
```
